# Simple example of ARIMA forecasting of electricity and gas production in the United States

```python
# Import all the needed packages

import numpy as np
import pandas as pd
from pandas import datetime
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Import and explore the data of industrial production of electricity and gas utilities in the United States
# The data is in monthly frequency ranging from 1985 to 2020

data = pd.read_csv('/Users/kshit/Desktop/Python/ELECGAS.csv', index_col = 0)
data.index=pd.to_datetime(data.index)
data.columns = ['Production']
print(data.head())
```

```python
# Visualizing the production of the electricity and gas

data.plot(figsize = (12,8), title = 'Energy Production')
plt.xlabel('Date')
plt.ylabel('Index 2012 = 1000')
plt.show()
```

Seasonality is a characteristics of a time series in which regular and predictable changes can be seen recurring every calender year. It occurs at specific regular intervals such as weekly, monthly, or quarterly. Time series data may contain seasonal variation and it is important to understand and identify the seasonal component of a time series because it can help us to make our model more effecient.  

```python
# Decomposing the time series
# We can see the strong seasonality in the production of the utilities

result = seasonal_decompose(data, 'multiplicative')
result.plot()
plt.show()
```

In simple term, stationarity is a statistical property where mean, variance, autocorrelation, etc. are all constant over the time. 
It is very essential for us to check if the time series is stationary or not. One of the reason for time series to be stationary 
is to avoid the problem of spurious regression. 

We can check for stationarity through augmented Dickey-Fuller test (ADF). It tests the null hypothesis that a unit root is presesnt in time series.  

```python
# Determining if the time series is stationary or not
# Running dickey-fuller test on the time series 

result =adfuller(data['Production'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
# From the above test we conclude the time series is non stationary. 
# Taking the first difference of time series and droping NaN values

data_diff = data.diff().dropna()
result_diff = adfuller(data_diff['Production'])
print('ADF Statistic:', result_diff[0])
print('p-value:', result_diff[1])
```

```python
# Plotting the time series after the first difference

data_diff.plot(figsize = (12,8), title = 'Energy Production')
plt.show()
```

```python
# Splitting the data into train and test set
x=data.values
data_train = x[:350]
data_test = x[350:]
len(data_test)
```

# Building ARIMA Model

ARIMA (Autoregressive Integrated Moving Average) is a statistical model for analysing and forecasting time series data. ARIMA model contains three parameters.
    1. p = Autoregression order
    2. d = Order of difference needed for stationarity
    3. q = Moving Average order
    
For more detail please refer to: https://people.duke.edu/~rnau/411arim.htm

Both AIC (Akaike Information Criteria) and BIC (Bayesian Information Criteria) estimates the quality of model by penalizing the overfitting of the model. Model with less AIC and BIC is considered good model. 

https://machinelearningmastery.com/probabilistic-model-selection-measures/

```python
# Finding AIC and BIC 
# Searching the model order with lowest AIC and BIC

import warnings
warnings.filterwarnings('ignore')
for p in range(5):
    
    for q in range(5):
        
        try:
            model = ARIMA(data_train, order = (p,1,q))
            fit_model = model.fit()
            
            print(p,q,fit_model.aic, fit_model.bic)
       
        except:
            print(p,q,None,None)
```

```python
# Fitting the ARIMA model into the train data set.
# Forecasting the next 72 months production 
model = ARIMA(data_train, order = (4,1,2))
fit_model = model.fit()
forecasted_data = fit_model.forecast(steps = 72)[0]
forecasted_data
```

```python
# Plotting the forecasted data with the actaul test data

plt.figure(figsize=(15,10))
plt.plot(forecasted_data, color='red', label = 'Forecasted')
plt.plot(data_test, label = 'Observed')
plt.legend()
plt.show()
```

# SARIMA

SARIMA (Seasonal Autoregressive Integrated Moving Average) model is the direct extension of the ARIMA model. SARIMA supports
the direct modeling of the seasonal component of a time series. SARIMA model includes four more parameters. There are four seasonal elements that are not part of ARIMA that must be configured. They are:

    1. P = Seasonal Autoregressive order
    2. Q = Seasonal Moving Average order
    3. D = Seasonal Difference order
    4. m = The number of time steps for a single seasonal period
    
SARIMA model is specified as SARIMA(p,d,q)(P,D,Q)m.

https://otexts.com/fpp2/seasonal-arima.html

auto_arima function is one of the easiest and less complex approach of finding the best parameters for the SARIMA model. 
This function allows us to find the best combination of the parameters with the lowest AIC value. 

```python
# Import the pmdarima for auto_arima function. 

import pmdarima as pm
```

```python
# Run the function with the specified arguments to find the best model

sarima_model = pm.auto_arima(data, seasonal = True, m=12,d=1,D=1,start_p=1,start_q=1,start_P=0,max_p=3,max_q=3,
                             trace=True,error_action='ignore',
                            suppress_warnings=True, stepwise=True)
```

```python
# Fitting the model

sarima_model.fit(data_train)
```

```python
# Making the prediction based on the fitted SARIMA model

future_forecast = sarima_model.predict(n_periods=72)
```

```python
# Plotting the predicted data

plt.figure(figsize=(15,10))
plt.plot(future_forecast, color='red', label = 'Forecasted')
plt.plot(data_test, label = 'Observed')
plt.legend()
plt.show()
```

From the above figure, we can see that SARIMA model has done a better job of predicting the future production of electricity and 
gas utilities in compare to ARIMA model. 

Out final step is to perform model diagnostics. Our purpose is to find whether the model is good fit for our data or not.

```python
# plotting the model diagnostics

sarima_model.plot_diagnostics(figsize=(12,8))
plt.show()
```

From the above plot we can conclude the following points. 

   - Standard Residual: There are no obvious patterns in the residuals.
   - Histogram plus kde : The KDE curve is similar to normal distribution.
   - Normal Q-Q: Most of the data points lie on the straight line.
   - Correlogram: The residuals are not correlated.
   
Overall, the model seems to be a good fit.
